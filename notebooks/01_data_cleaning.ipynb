```python
import os
from pandas import DataFrame
import pandas as pd
from pathlib import Path
```

```python
def clean_study(study_annotation, pipeline_version, group_type='GO_slim'):
    '''
    merge all sample go_slim result and split into each description. returns a csv files in ../data/{study_annotation}/{pipeline_version}
    '''
    path = Path(f'../data/{study_annotation}/{pipeline_version}')
    targets = [i for i in os.listdir(path / 'functional_analysis') if i.endswith(f'{group_type}.csv')]
    samples = [i.split('/')[-1].split('_')[0] for i in targets]
    df_cum = []
    for num, i in enumerate(targets):
        df = pd.read_csv(path / 'functional_analysis' / i, header=None)
        df.columns = ['GO_term', 'category', 'description', samples[num]]
        df = df.set_index(['GO_term', 'category', 'description'])
        df_cum.append(df)
    df_fin = pd.concat(df_cum, axis=1, join="inner")
    index_cat = df_fin.index.get_level_values(2).unique()
    df_split = [df_fin[df_fin.index.isin([i], level=2)] for i in index_cat]
    for num, i in enumerate(df_split):
        i.to_csv(path / f'{study_annotation}_{pipeline_version}_{group_type}_{index_cat[num]}.csv')
    return #df_split
```

```python
# get available datasets
path = Path('../data')
study_annotations = [item for item in os.listdir(path) if os.path.isdir(os.path.join(path, item))]
study_annotations.remove('.ipynb_checkpoints')

pipeline_version = '4.1'
#group_type = 'GO_slim'

# merge and split
for i in study_annotations:
    clean_study(i, pipeline_version)
```

```python
group_types = ['biological_process', 'cellular_component', 'molecular_function']
for g in group_types:
    df_cum = []
    for i in study_annotations:
        path = Path('../data/') / i / pipeline_version / f'{i}_{pipeline_version}_GO_slim_{g}.csv'
        df = pd.read_csv(path)
        df = df.set_index(['GO_term', 'category', 'description'])
        df_cum.append(df)
    df_fin = pd.concat(df_cum, axis=1, join="inner")
    df_fin.to_csv(f'../data/GO_slim_all_studies_{g}.csv')
```
