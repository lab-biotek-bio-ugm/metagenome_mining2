```python
import os
from pandas import DataFrame
import pandas as pd
from pathlib import Path
```

```python
path = Path('../data/PRJEB26427/4.1/functional_analysis/')


targets = [i for i in os.listdir(path) if i.endswith('_FASTQ_GO_slim.csv')]
samples = [i.split('/')[-1].split('_')[0] for i in targets]
df_cum = []
for num, i in enumerate(targets):
    df = pd.read_csv(path / i, header=None)
    df.columns = ['GO_term', 'category', 'description', samples[num]]
    df = df.set_index(['GO_term', 'category', 'description'])
    df_cum.append(df)
df_fin = pd.concat(df_cum, axis=1, join="inner")
```
